```python
import numpy as np

# ANSWERS FOR T = 16 
# Number of moves
T = 16

# These variables help identify movable keys on the board for each initial position 
moving_distance = [1,-1,2,-2]
limit = range(0,4)
movable_key = []

# This array handles the calculated key sum for each iteration
S_handle = []

# Number of iterations
num_of_iter = 100000

for iterator in range(0,num_of_iter):
    initial_key = 0
    S = initial_key    
    for t in range(0,T):
        # identify movable keys
        for i in moving_distance:
            for j in moving_distance:
                if ((abs(i) + abs(j) == 3) & ((initial_key%4 + i) in limit) & ((initial_key//4 + j) in limit)):
                    movable_key.append(initial_key + i + 4*j)
        movable_key = np.array(movable_key) 
        # choose the next move in a uniformly random manner
        a = np.random.randint(movable_key.size,size=1) 
        next_move_key = movable_key[a[0]]
        # calculate key sum
        S = S + next_move_key
        initial_key = next_move_key
        movable_key = []
    # construct the key sum handle
    S_handle.append(S)

# Calculate mean modulo 13 and variance of key sum 
S_handle = np.array(S_handle) 
mean_mod_13 = S_handle.mean()%13
var = S_handle.std()
print('%.10f' % mean_mod_13,'%.10f' % var) # final result

# Calculate conditional probability of key sum divisible by 5 given that it is divisible by 13
S_13_count = 0
S_5_13_count = 0
for i in range(0,S_handle.size):
    if S_handle[i]%13 == 0:
        S_13_count = S_13_count + 1
        if S_handle[i]%5 == 0:
            S_5_13_count = S_5_13_count + 1
prop = S_5_13_count/S_13_count*100.0
print('%.10f' % prop) # final result


        
# ANSWER FOR T = 512
# Number of moves
T = 512

# These variables help identify movable keys on the board for each initial position 
moving_distance = [1,-1,2,-2]
limit = range(0,4)
movable_key = []

# This array handles the calculated key sum for each iteration
S_handle = []

# Number of iterations
num_of_iter = 100000

for iterator in range(0,num_of_iter):
    initial_key = 0
    S = initial_key    
    for t in range(0,T):
        # identify movable keys
        for i in moving_distance:
            for j in moving_distance:
                if ((abs(i) + abs(j) == 3) & ((initial_key%4 + i) in limit) & ((initial_key//4 + j) in limit)):
                    movable_key.append(initial_key + i + 4*j)
        movable_key = np.array(movable_key) 
        # choose the next move in a uniformly random manner
        a = np.random.randint(movable_key.size,size=1) 
        next_move_key = movable_key[a[0]]
        # calculate key sum
        S = S + next_move_key
        initial_key = next_move_key
        movable_key = []
    # construct the key sum handle
    S_handle.append(S)

# Calculate mean modulo 311 and variance of key sum 
S_handle = np.array(S_handle) 
mean_mod_311 = S_handle.mean()%311
var = S_handle.std()
print('%.10f' % mean_mod_311, '%.10f' % var) # final result

# Calculate conditional probability of key sum divisible by 7 given that it is divisible by 43
S_43_count = 0
S_7_43_count = 0
for i in range(0,S_handle.size):
    if S_handle[i]%43 == 0:
        S_43_count = S_43_count + 1
        if S_handle[i]%7 == 0:
            S_7_43_count = S_7_43_count + 1
prop = S_7_43_count/S_43_count*100.0
print('%.10f' % prop) # final result
        

    
    

    
```
